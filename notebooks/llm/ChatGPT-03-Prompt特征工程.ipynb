- 步骤1：文本编码TFIDF对文本和知识库编码
- 步骤2：问题与知识库中最相关的5个新闻
- 步骤3：构建prompt提问，通过API得到回答



````python
import numpy as np
documents = open('dataset/爬虫-新闻标题.txt').readlines()
documents[0] = '武大发生了什么事情？'
documents = np.array(documents)


from sklearn.feature_extraction.text import TfidfVectorizer
import jieba
import numpy as np
segmented_sentences = [" ".join(jieba.cut(sentence)) for sentence in documents]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(segmented_sentences)
tfidf_encoded = tfidf_matrix.toarray()



retrieval_results = documents[
    np.dot(tfidf_encoded[0], tfidf_encoded.T).argsort()[::-1][1:4]
]
retrieval_results = ''.join(retrieval_results)

prompt_text = f'''
请帮我回答：武大最近发生了事情？

已知信息为：
1、在12月，武大有马拉松
2、在12月，有期末考试.
'''
````
